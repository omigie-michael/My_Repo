### The Code below is going to give us a model through which we can successfully convert weight from Kg to bs and vise versa.

- First we'll input the weight in Kg and get the conversion in bs
- Secondly, we'll input the weight in bs and get our conversion in Kg

```python
weight = input ("weight: ")
unit = input ("(K)Kg or (L)bs: ")
if unit == "K":
    converted = (float(weight)/0.45)
    print ("weight in bs: " + str(converted))
else:
    reverted = (float(weight) * 0.45)
    print ("weight in Kg: " + str(reverted))
```
